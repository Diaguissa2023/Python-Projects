<b>Les variables quantitatives </b>
<p> A l'inverse, ces variables permettent de mesurer des quantités <br>
Il en existe 2 types : Variables quantitatives discrètes et continues </p>

<b>Les variables quantitatives discrètes </b>
<p> C'est une variable avec un nombre fini de valeurs <br>
Exemples : nombre d'items, nombre de personnes <p>

```python
#les variables quantitatives discrètes
sns.distplot(list(data['Memory']))
```

```python
sns.distplot(list(data['Lv 50 HP']))
```

<b>Les variables quantitatives continues </b>

Exemples : poids, durée <p>

<b> Description d'une variable continue </b> <br>
<b> La moyenne </b>
<p> Somme des valeurs de la variable divisée par le nombre d'individus </p>



```python
mean = data['Lv 50 HP'].mean()
print(mean)
sns.distplot(list(data['Lv 50 HP']))
plt.axvline(mean, color='r')
```

<b> La Médiane </b>
<p>  valeur centrale qui partage l'échantillon en 2 groupes de même effectif : 50 % au-dessus et 50 % en dessous <p>

```python
median = data['Lv 50 HP'].median()
print(median)
sns.distplot(list(data['Lv 50 HP']))
plt.axvline(mean, color='r')
plt.axvline(median, color='g')
```

<b> Le mode </b>
<p> C'est la valeur la plus fréquente </p>

```python
mode = data['Lv 50 HP'].mode()
print(mode)
sns.distplot(list(data['Lv 50 HP']))
plt.axvline(mean, color='r')
plt.axvline(median, color='g')
plt.axvline(float(mode), color='y')
```

```python
mode = data['Lv 50 HP'].mode()
print(mode)
nbValDif = len(data['Lv 50 HP'].unique())
sns.distplot(list(data['Lv 50 HP']), bins=nbValDif)
plt.axvline(float(mode), color='y')
```

```python
mode = data['Type'].mode()
print(mode)
sns.countplot(x="Type", data=data)
```

<b> Calculer le coéficient d'assymétrie (skewness) de sa distribution </b>

Fait référence à l'asymétrie des données. Lorsque les données sont concentrées sur le côté droit de l'histogramme, par exemple, nous disons qu'elles ont un skew négatif. Lorsque les données sont concentrées à gauche, on dit qu'elles ont un skew positif.

```python
sns.distplot(list(data['Lv 50 HP']))
data['Lv 50 HP'].skew()
```

<b> Calculer le coéficient d'applatissement d'une distribution : Kurtosis </b>

<p>Plus la valeur d'aplatissement est élevée, plus le pic est élevé. Plus la valeur de kustose est faible, plus la valeur est étalée.</p>



```python
sns.distplot(list(data['Lv 50 HP']))
data['Lv 50 HP'].kurtosis()
```
