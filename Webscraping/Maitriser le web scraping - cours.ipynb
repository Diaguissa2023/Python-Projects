<b>Recupérer des informations depuis une page web</b> <br>

Beautiful Soup est une bibliothèque Python pour extraire des données de fichiers HTML et XML. Il fonctionne avec votre analyseur favori pour fournir des moyens simplifiés de navigation, de recherche et de modification de l'arbre d'analyse. Il permet généralement aux programmeurs d'économiser des heures ou des jours de travail.

```python
from bs4 import BeautifulSoup
html = BeautifulSoup(page.text, 'html.parser')
html
```

```python
html.find('h1')
#trouver le titre de la page
```

```python
html.find('h1').getText()
```

```python
html.find('body').getText()
```

```python
html.find_all('p')
```

```python
#objectif : récuperer les articles des plus lus
html.find_all('a')
```

```python
html.find_all('span',{'class':'insert__subtitle'})
```

```python
topArticles = html.find_all('span',{'class':'insert__subtitle'})
topArticlesClean = [x.getText() for x in topArticles]
topArticlesClean
```
